## Loading My Data

```python
# We have been given a data set which is in a csv (comma-separatedvalues) file.
# First and foremost we have to read the dataset in a pandas dataframe and to do this we have steps which will be listed below.
# A dataframe is a 2 Dimensional table.

# To be able to read our dataset into pandas we have to;
# A. upload the dataset into our jupyter notebook folder,
# so it can be accessesible by the computer without throwing an error of file not found
# B. import pandas under any alias mine i choose pd, yours you can choose pnd, e.t.c.
# Pandas is a software library written for the Python programming language for data manipulation and analysis.

import pandas as pd
```

```python
def read_dataset(file_path):

    """
    This function reads a dataset from a file into a pandas DataFrame and handles errors gracefully.

    Parameters:
    file_path (str): The path to the dataset file.

    Returns:
    pd.DataFrame or None: Returns the DataFrame if successful, otherwise None.
    """
    try:
        # Try reading the file
        data_set = pd.read_csv(file_path)
        print("File loaded successfully.")
        return data_set
    except FileNotFoundError:
        # Handle the case where the file is not found
        print(f"Error: The file at path '{file_path}' was not found.")
    except pd.errors.EmptyDataError:
        # Handle the case where the file is empty
        print(f"Error: The file at path '{file_path}' is empty.")
    except pd.errors.ParserError:
        # Handle the case where the file contents are not parseable
        print(f"Error: The file at path '{file_path}' could not be parsed. Please check if it's a valid CSV.")
    except Exception as e:
        # Catch all other exceptions and print a generic message
        print(f"An unexpected error occurred: {str(e)}")
    
    # Return None if an error occurred
    return None
```

## FEATURE ENGINEERING

```python
# Feature Engineering to create useful columns inside the dataset (Data Preparation)
# Data preparation to show Cost Efficiency features

data_set["Cost_Efficiency"] = (data_set["Cost"] / (data_set["Distance_miles"] * data_set["Weight_kg"])).round(2)
```

```python
data_set.tail(12)
```

```python
# Route features to help analyze route performance
data_set["Route"] = data_set["Origin_Warehouse"] + " - " + data_set["Destination"]
```

```python
data_set.tail(12)
```

```python
# Delay Days feature shows us how late or early deliveries arrived
data_set["Delay_Days"] = (data_set["Delivery_Date"] - data_set["Shipment_Date"]).dt.days - data_set["Transit_Days"]
```

```python
data_set.head(12)
```

```python
# Is Delay feature makes the data set more presentable in the sense that wherever there is delay it should all show 1 and where there isnt delay should remain as 0
# It doesnt matter if the delay days is more than 1 we just using 1 to denote that there was a delay in this particular delivery
data_set["Is_Delayed"] = data_set["Delay_Days"].apply(lambda x: 1 if x > 0 else 0)
```

```python
data_set.iloc[1025:1058]
```

```python
# Shipping Speed measures efficiency of movement
data_set["Shipping_Speed"] = (data_set["Distance_miles"] / data_set["Transit_Days"]).round(2)
```

```python
data_set.head(12)
```

```python
data_set[["Delay_Days", "Is_Delayed", "Shipping_Speed"]].describe()
```

```python
data_set["Is_Delayed"].value_counts()
```

```python
# Shipment Month feature to be able to detect season
data_set["Shipment_Month_Num"] = data_set["Shipment_Date"].dt.month
data_set["Shipment_Month"] = data_set["Shipment_Date"].dt.month_name()
```

```python
data_set.head(15)
```

```python
data_set["Carrier_Code"] = data_set["Carrier"].astype("category").cat.codes
```

```python
data_set.head(15)
```

```python
# Distance Category feature helps us to see the catergory of the distance in each trip
data_set["Distance_Category"] = pd.cut(
    data_set["Distance_miles"],
    bins = [0, 300, 1000, data_set["Distance_miles"].max()],
    labels = ["Short", "Medium", "Long"]
)
```
